```python
import numpy as np
import matplotlib.pyplot as plt

def calc_V_rate(epsilon, D, V_rateall):
  for d in range(D.shape[0]):
      V_rate = []
      for e in range(epsilon.shape[0]):
        result = 1-np.power((1-epsilon[e]),D[d])
        V_rate.append(result)
      V_rateall.append(V_rate)

def plot_V_rate(epsilon, V_rateall):
    fig = plt.figure()
    for d in range(D.shape[0]):
      plt.plot(epsilon,V_rateall[d],label="D="+str(D[d]))

    plt.xlabel("$\epsilon$")
    plt.ylabel("Rate of Volume" )
    plt.legend()
    plt.savefig("hypersphere_sample.png", format='png', dpi=500)

if __name__=="__main__":

    epsilon = np.arange(0, 1, 0.01)
    D = np.array([1, 2, 3, 6, 11, 16, 21, 100, 500, 1000])
    V_rateall = []

    calc_V_rate(epsilon, D, V_rateall)
    plot_V_rate(epsilon, V_rateall)
```
